**2.a.** Los valores para `Pobres`, `Vulnerables`, `Clase_Media`, `Clase_Alta` están en habitantes, calcule esos valores en porcentaje y agreguelos a sus datos.

```python
# Lista para: Totalidad Personas Dpto
totalHabitantes = []

# LLenar en totalHabitantes la totalidad de habitantes  de cada dpto.
for i in range(0, len(lista)):
    totalHabitantes.append((lista[i]['Pobres']) + (lista[i]['Vulnerables']) + (lista[i]['Clase_Media']) + (lista[i]['Clase_Alta']))

#Imprimir la columnas y sus valores 
    
print(" - Pobres - ")
for recorrer in range(0, len(lista)):
    print("Departamento --> {}:  Pobres --> {}".format(lista[recorrer]['Departamento'], lista[recorrer]['Pobres']))
    
   
print() 

print(" - Vulnerables - ")
for recorrer in range(0, len(lista)):
    print("Departamento --> {}:  Vulnerables --> {}".format(lista[recorrer]['Departamento'], lista[recorrer]['Vulnerables']))
    

   
print() 

print(" - Clase Media - ")
for recorrer in range(0, len(lista)):
    print("Departamento --> {}:  Clase Media --> {}".format(lista[recorrer]['Departamento'], lista[recorrer]['Clase_Media']))
   
print() 

print(" - Clase Alta - ")
for recorrer in range(0, len(lista)):
    print("Departamento --> {}:  Clase Alta --> {}".format(lista[recorrer]['Departamento'], lista[recorrer]['Clase_Alta']))
   
print() 

print("Total de habitantes")
print(totalHabitantes)


#Agregando las columnas de porcentajes pobres, vulnerables, clase media, clase alta
for recorrer in range(0, len(lista)):
    
    porcentajePobres= ((lista[recorrer]['Pobres'])*100/(totalHabitantes[recorrer]))/100
    lista[recorrer]['PorcentajePobres'] = porcentajePobres
    
    porcentajeVulnerables = ((lista[recorrer]['Vulnerables'])*100/(totalHabitantes[recorrer]))/100
    lista[recorrer]['PorcentajeVulnerables'] = porcentajeVulnerables
    
    porcentajeClaseMedia = ((lista[recorrer]['Clase_Media'])*100/(totalHabitantes[recorrer]))/100
    lista[recorrer]['PorcentajeClaseMedia'] = porcentajeClaseMedia
    
    porcentajeClaseAlta = ((lista[recorrer]['Clase_Alta'])*100/(totalHabitantes[recorrer]))/100
    lista[recorrer]['PorcentajeClaseAlta'] = porcentajeClaseAlta
    
```

## Respuesta


**2.b.** Los valores para `Porc_Edad_Trabajar`, `Porc_Econ_Activa`, `Porc_Ocupados`, `Porc_Desocupados` están en porcentajes, calcule esos valores en habitantes y agreguelos a sus datos.

## Respuesta

```python
for recorrer in range(0, len(lista)):
    
    habitantes_Porc_Edad_Trabajar = lista[recorrer]['Porc_Edad_Trabajar'] * totalHabitantes[recorrer]
    lista[recorrer]['Habitantes_Porc_Edad_Trabajar'] = habitantes_Porc_Edad_Trabajar
    
    habitantes_Porc_Econ_Activa = lista[recorrer]['Porc_Econ_Activa'] * totalHabitantes[recorrer]
    lista[recorrer]['Habitantes_Econ_Activa'] = habitantes_Porc_Econ_Activa
    
    habitantes_Porc_Ocupados = lista[recorrer]['Porc_Ocupados'] * totalHabitantes[recorrer]
    lista[recorrer]['Habitantes_Porc_Ocupados'] = habitantes_Porc_Ocupados
    
    habitantes_Porc_Desocupados = lista[recorrer]['Porc_Desocupados'] * totalHabitantes[recorrer]
    lista[recorrer]['Habitantes_Porc_Desocupados'] = habitantes_Porc_Desocupados
    
    
lista
```
